## Uniform Distribution

```python
n = 1000
start = 10
width = 20
rv_uniform = uniform.rvs(size=n, loc=start, scale=width)
rv_uniform
```

```python
sns.distplot(
    rv_uniform,
    bins=100,
    kde=True,
    color='skyblue',
    hist_kws={'linewidth': 15, 'alpha':1}
)
plt.xlabel('uniform dist')
plt.ylabel('Frequency')
plt.show()
```

## Normal Distribution

```python
n=100
avg = 0
stdev = 1
rv_normal = norm.rvs(size=n, loc=avg, scale=stdev)
rv_normal
```

```python
sns.distplot(
    rv_normal,
    bins=100,
    kde=True, #kernel density estimation
    color='skyblue',
    hist_kws={'linewidth': 15, 'alpha':1}
)
plt.xlabel('standard normal distribution')
plt.ylabel('frequency')
plt.show()
```

## Confidence Interval

```python
df = pd.read_csv('heart.csv')
print(df.shape)
df.head()
```

```python
df['sex_name'] = df.sex.replace({1: 'male', 0:'female'})
df.head()
```

```python
df.info()
```

```python
dx = df[['sex_name', 'target']]
dx
```

```python
pd.crosstab(dx.target, dx.sex_name)
```

```python
pop_female = 86 + 226
prop_female = 226/pop_female
print('female population:', pop_female)
print('proportion of female having disease:', prop_female)
```

```python
stderr_female = np.sqrt(
    prop_female*(1-prop_female)/ pop_female
)
print('standard error for proportion:', stderr_female)
```

```python
def compute_confint(estimate, stderr, z_score=1.96):
    return(
        estimate - z_score*stderr,
        estimate + z_score*stderr
    )
```

```python
compute_confint(prop_female, stderr_female)
```

```python
z_score = 1.96
lower_ci = prop_female - (z_score*stderr_female)
upper_ci = prop_female + (z_score*stderr_female)
print('CI: [{}, {}]'.format(lower_ci, upper_ci))
```

## Hypothesis Test

```python
n = 1018
pnull = .52
phat = .56
```

```python
sm.stats.proportions_ztest(
    phat*n,
    n,
    pnull,
    alternative='larger'
)
#p-val <0.05 reject hipotesis
```
