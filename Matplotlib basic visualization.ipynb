```python
years = list(range(1980, 2014))
year80s = list(range(1980, 1990))
year90s = list(range(1990, 2000))
year00s = list(range(2000, 2010))
```

```python
haiti = df_canada.loc[df_canada['Country']=='Haiti', years].transpose()
haiti.rename(columns={75: 'Haiti'}, inplace=True)
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot()
plt.ylabel('Number of Immigration')
plt.xlabel('Years')
plt.title('Immigration from Haiti over Years')
#plt.legend('')

plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```

```python
cn_in = df_canada.loc[
    df_canada['Country'].isin(['China', 'India']),
    years
].transpose()
cn_in.rename(
    columns={
        36: 'China',
        79: 'India'
    },
    inplace=True
)
cn_in.head()
```

```python
cn_in.plot()

plt.title('Immigrant China & India in Canada over Years')
plt.xlabel('Years')
plt.ylabel('Number of Immigrant')
plt.show()
```

```python
#SORT DATA BY TOTAL
top5_countries = df_canada.sort_values(
        by='Total', ascending=False, axis=0
).head()
top5countries = top5_countries.set_index('Country')
top5countries = top5countries[years].transpose()
```

```python
top5countries.head()
```

```python
top5countries.plot(figsize=(15, 10))
plt.title('Top 5 Immigrant Origin in Canada over Years')
plt.ylabel('Num of Immigrant')
plt.xlabel('Years')
plt.show()
```

```python
#AREA PLOT
#plt.style bla bla
top5countries.plot(
    kind='area',
    stacked=False,
    figsize=(20,10)
)
plt.title('Top 5 Immigrant Origin in Canada over Years')
plt.ylabel('Num of Immigrant')
plt.xlabel('Years')
plt.show()
```

```python
top5countries.plot(
    kind='area',
    alpha=0.25,
    stacked=True,
    figsize=(20,10)
)
plt.title('Top 5 Immigrant Origin in Canada over Years')
plt.ylabel('Num of Immigrant')
plt.xlabel('Years')
plt.show()
```

```python
ax = top5countries.plot(
    kind='area',
    alpha=0.25,
    stacked=False,
    figsize=(20,10)
)
ax.set_title('Top 5 Immigrant Origin in Canada over Years')
ax.set_ylabel('Num of Immigrant')
ax.set_xlabel('Years')
plt.show()
```

```python
least5_countries = df_canada.sort_values(
        by='Total', ascending=True, axis=0
).head()
least5countries = least5_countries.set_index('Country')
least5countries = least5countries[years].transpose()
```
